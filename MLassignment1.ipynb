```python

X = [1, 2, 3]
Y = [2, 4, 6]


m = 0.5
c = 0.5
alpha = 0.1


for iteration in range(6):
    n = len(X)

    dm = 0
    dc = 0
    for i in range(n):
        y_pred = m * X[i] + c
        dm += -(Y[i] - y_pred) * X[i]
        dc += -(Y[i] - y_pred)
    dm /= n
    dc /= n


    m = m - alpha * dm
    c = c - alpha * dc


    print(f"Iteration {iteration + 1}: m = {m:.4f}, c = {c:.4f}")


print("\nFinal values:")
print(f"m = {m:.4f}, c = {c:.4f}")
```
